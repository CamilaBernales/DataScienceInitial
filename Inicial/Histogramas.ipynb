```python
# seaborn graficos estadisticos
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib as mpl
```

```python
%matplotlib inline
```

```python
datos1 = np.random.randn(100)
datos1
```

```python
mpl.pyplot.hist(datos1)
```

```python
sns.distplot(datos1)
```

```python
sns.distplot(datos1, color="pink")
```

```python
mpl.pyplot.hist(datos1, color="pink", alpha=1)
```

```python
datos2 = np.random.randn(80)
datos2
```

```python
mpl.pyplot.hist(datos2, color="yellow")
```

```python
# juntamos los datos

mpl.pyplot.hist(datos1, color="green", alpha= 0.3, bins=20, density=True) # bin numeros de barras
mpl.pyplot.hist(datos2, color="blue", alpha= 0.3, bins=20, density=True) # bin numeros de barras
```

```python
datos3 = np.random.randn(1000)
datos4 = np.random.randn(1000)
```

```python
sns.jointplot(datos3, datos4)
```

```python
sns.jointplot(datos3, datos4, kind="hex")
```
