```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
```

```python
datos = np.random.randn(100)
datos
```

```python
sns.distplot(datos, color='green')
```

```python
sns.distplot(datos, color='green', rug=False, hist=False)
```

```python
arg_curva = {'color':'black', 'label':'Curva'}
arg_hist = {'color':'gray', 'label':'Histograma'}
sns.distplot(datos, bins=25, kde_kws=arg_curva, hist_kws=arg_hist)
```

```python
serie = pd.Series(datos)
serie
```

```python
sns.distplot(serie, bins=25, color="green")
```
