```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
# mapas de calor => representacion graficas, donde los valores individuales contenidos en una matriz se representan como colores
```

```python
vuelos = sns.load_dataset('flights')
vuelos.head(20)
```

```python
vuelos = vuelos.pivot_table('month', 'year', 'passengers')
vuelos
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos, annot=True, fmt='d')
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
```

```python
sns.heatmap(vuelos, center = valor_central, annot=True, fmt="d")
```

```python
sns.heatmap(vuelos, center = valor_central, annot=True, fmt="d", cbar_kws={'orientation':'horizontal'})
```
